### Split into test and train

```python
#break out features and class
X_lvl1 = level1df.drop(['isSpoofed'], axis=1)
y_lvl1 = level1df['isSpoofed']

X_lvl2 = level2df.drop(['isSpoofed'], axis=1)
y_lvl2 = level2df['isSpoofed']

X_lvl3 = level3df.drop(['isSpoofed'], axis=1)
y_lvl3 = level3df['isSpoofed']

X_lvl4 = level4df.drop(['isSpoofed'], axis=1)
y_lvl4 = level4df['isSpoofed']

X_lvl5 = level5df.drop(['isSpoofed'], axis=1)
y_lvl5 = level5df['isSpoofed']

#X_combined = combineddf.drop(['isSpoofed'], axis=1)
#y_combined = combineddf['isSpoofed']
```

```python
#get test and train sets
Xlvl1_train, Xlvl1_test, ylvl1_train, ylvl1_test = train_test_split(X_lvl1, y_lvl1, test_size=0.2, random_state=42)
Xlvl2_train, Xlvl2_test, ylvl2_train, ylvl2_test = train_test_split(X_lvl2, y_lvl2, test_size=0.2, random_state=42)
Xlvl3_train, Xlvl3_test, ylvl3_train, ylvl3_test = train_test_split(X_lvl3, y_lvl3, test_size=0.2, random_state=42)
Xlvl4_train, Xlvl4_test, ylvl4_train, ylvl4_test = train_test_split(X_lvl4, y_lvl4, test_size=0.2, random_state=42)
Xlvl5_train, Xlvl5_test, ylvl5_train, ylvl5_test = train_test_split(X_lvl5, y_lvl5, test_size=0.2, random_state=42)
#Xcombined_train, Xcombined_test, ycombined_train, ycombined_test = train_test_split(X_combined, y_combined, test_size=0.2, random_state=42)
```

```python
Xcombined_train = Xlvl1_train
Xcombined_train.append(Xlvl2_train)
Xcombined_train.append(Xlvl3_train)
Xcombined_train.append(Xlvl4_train)
Xcombined_train.append(Xlvl5_train)

Xcombined_test = Xlvl1_test
Xcombined_test.append(Xlvl2_test)
Xcombined_test.append(Xlvl3_test)
Xcombined_test.append(Xlvl4_test)
Xcombined_test.append(Xlvl5_test)

ycombined_train = ylvl1_train
ycombined_train.append(ylvl2_train)
ycombined_train.append(ylvl3_train)
ycombined_train.append(ylvl4_train)
ycombined_train.append(ylvl5_train)

ycombined_test = ylvl1_test
ycombined_test.append(ylvl2_test)
ycombined_test.append(ylvl3_test)
ycombined_test.append(ylvl4_test)
ycombined_test.append(ylvl5_test)
```
